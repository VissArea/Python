```python
height = [3,12,5,18,45]
bars = ('A', 'B', 'C', 'D', 'E')
y_pos = np.arange(len(bars))

#Create bars
plt.bar(y_pos, height)

#Create X-axis names
plt.xticks(y_pos, bars)

#Show Plot
plt.show()
```

```python
height = [3,12,5,18,45]
bars = ('A', 'B', 'C', 'D', 'E')
explode = (0, 0.1, 0, 0, 0) #Makes the B section explode out of the Pie graph

fig1, ax1 = plt.subplots()
ax1.pie(height, explode=explode, labels=bars, autopct='%1.1f%%', 
        shadow=True, startangle=90)
ax1.axis('equal') #Equal aspect Ratio ensure that pie is drawn in a circle
plt.show()
```

```python
def hist_and_lines():
  np.random.seed(0)
  fig, ax = plt.subplots(1,2,figsize=(11,4))
  ax[0].hist(np.random.rand(1000))
  for i in range(3):
    ax[1].plot(np.random.rand(10))
  ax[1].legend(['A', 'B', 'C'], loc='lower left')
```

```python
hist_and_lines()
```

```python
with plt.style.context('fivethirtyeight'):
  hist_and_lines()
```

```python
with plt.style.context('ggplot'):
  hist_and_lines()
```
